```python
import numpy as np
import matplotlib.pyplot as plt

# 랜덤 시드 생성(역자 추가한 코드 한 줄)
np.random.seed(42)

# 샘플 데이터 생성
x = np.random.randn(5, 10) # 5개의 샘플, 10차원의 데이터
g = np.random.randn(1, 10) # 메인 매개 변수
b = np.random.randn(1, 10) # 바이어스 매개 변수

# layer_norm 함수 정의
def layer_norm(x, g, b, eps=1e-5):
  mean = np.mean(x, axis=-1, keepdims=True)
  variance = np.var(x, axis=-1, keepdims=True)
  return g * (x - mean) / np.sqrt(variance + eps) + b

# 정규화 전후의 데이터 플롯
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# 정규화하지 않은 원본 데이터
for i in range(x.shape[0]):
  axs[0].plot(x[i], label=f'Sample {i+1}')

axs[0].set_title('Before Layer Normalization')
axs[0].legend()

# 정규화한 데이터
x_norm = layer_norm(x, g, b)
for i in range(x_norm.shape[0]):
  axs[1].plot(x_norm[i], label=f'Sample {i+1}')

axs[1].set_title('After Layer Normalization')
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# 랜덤 시드 생성(역자 추가 코드)
np.random.seed(42)

# GELU와 Linear 함수 정의
def gelu(x):
  return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)))

def linear(x, w, b):
  return x @ w + b

# 피드 포워드 네트웤크(ffn) 함수 정의
def ffn(x, c_fc, c_proj):
  return linear(gelu(linear(x, **c_fc)), **c_proj)

# 샘플 데이터 생성
x = np.random.randn(2, 4)  # 입력 데이터(배치 크기 2, 차원 수 4)
w1 = np.random.randn(4, 8) # 1차 선형 층(레이어) 가중치 행렬(입력차원 4, 출력 차원 8)
b1 = np.random.randn(8)    # 1차 선형 층(레이어) 바이처스 벡터
w2 = np.random.randn(8, 4) # 2차 선형 층(레이어) 가중치 행렬(입력차원 8, 출력 차원 4)
b2 = np.random.randn(4)    # 2차 선형 층(레이어) 바이처스 벡터

# c_fc 및 c_proj 의 딕셔너리 작성
c_fc = {'w': w1, 'b': b1}
c_proj = {'w': w2, 'b': b2}

# ffn 함수 실행
out = ffn(x, c_fc, c_proj)

print("Input:\n", x)
print("Output:\n", out)
```

```python
import numpy as np

# Query、Key、Value의 더미 데이터
q = np.array([[1, 0, 1], [0, 1, 0]]) # (2, 3) Query 벡터
k = np.array([[0, 1, 0], [1, 0, 1], [0, 0, 1]]) # (3, 3) Key 벡터
v = np.array([[2, 3, 4], [5, 6, 7], [8, 9, 10]]) # (3, 3) Value 벡터

# mask의 더미 데이터 (모든 값이 0)
mask = np.zeros((2, 3))

def softmax(x):
  exp_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
  return exp_x / np.sum(exp_x, axis=-1, keepdims=True)

def attention(q, k, v, mask):
  return softmax(q @ k.T / np.sqrt(q.shape[-1]) + mask) @ v

print("Query:\n", q)
print("Key:\n", k)
print("Value:\n", v)
print("Mask:\n", mask)
print("Score:\n", q @ k.T / np.sqrt(q.shape[-1]))
print("Attention Weights:\n", softmax(q @ k.T / np.sqrt(q.shape[-1]) + mask))
print("Output:\n", attention(q, k, v, mask))

mask[0, 1] = -1e9 # 첫 번째 행의 두 번째 값을 마스크합니다.
print("Mask:\n", mask)
print("Attention Weights(Masked):\n", softmax(q @ k.T / np.sqrt(q.shape[-1]) + mask))
print("Output(Masked):\n", attention(q, k, v, mask))
```

```python
def mha(x, c_attn, c_proj, n_head):
  x = linear(x, **c_attn)
  qkv_heads = list(map(lambda x: np.split(x, n_head, axis=-1), np.split(x, 3,axis=-1)))
  causal_mask = (1 - np.tri(x.shape[0])) * -1e10
  out_heads = [attention(q, k, v, causal_mask) for q, k, v in zip(*qkv_heads)]
  x = linear(np.hstack(out_heads), **c_proj)
  return x
```

```python
def mha(x, c_attn, c_proj, n_head):
  batch_size, seq_len, dim = x.shape
  head_dim = dim // n_head

  x = linear(x, **c_attn)
  qkv = np.split(x, 3, axis=-1)
  q, k, v = map(lambda t: t.reshape(batch_size, seq_len, n_head, head_dim).transpose(0, 2, 1, 3), qkv)

  causal_mask = (1 - np.tri(seq_len, seq_len)) * -1e10
  causal_mask = causal_mask[None, None, :, :]

  q_prime = q.reshape(batch_size * n_head, seq_len, head_dim)
  k_prime = k.reshape(batch_size * n_head, seq_len, head_dim).transpose(0, 2, 1)

  scores = q_prime @ k_prime / np.sqrt(head_dim)
  scores = scores.reshape(batch_size, n_head, seq_len, seq_len)
  scores = scores + causal_mask
  attn_weights = softmax(scores)

  out = (attn_weights @ v).transpose(0, 2, 1, 3).reshape(batch_size, seq_len, dim)
  out = linear(out, **c_proj)

  return out

import numpy as np

# 샘플 데이터 생성
batch_size = 2
seq_len = 4
n_head = 2
dim = 8

# 랜덤 시드 추가 (역자가 추가한 한 줄 코드)
np.random.seed(42)

x = np.random.randn(batch_size, seq_len, dim)
c_attn = {'w': np.random.randn(dim, 3 * dim),'b': np.random.randn(3 * dim)}
c_proj = {'w': np.random.randn(dim, dim), 'b': np.random.randn(dim)}

# 멀티 헤드 어텐션 함수인 mha() 함수 실행
output = mha(x, c_attn, c_proj, n_head)

print(output.shape)
```

```python
def transformer_block(x, mlp, attn, ln_1, ln_2, n_head):
  x = x + mha(layer_norm(x, **ln_1), **attn, n_head=n_head)
  x = x + ffn(layer_norm(x, **ln_2), **mlp)
  return x

def gpt2(inputs, wte, wpe, blocks, ln_f, n_head):
  x = wte[inputs] + wpe[range(len(inputs))]
  for block in blocks:
    x = transformer_block(x, **block, n_head=n_head)
  return layer_norm(x, **ln_f) @ wte.T
```

```python
def generate(inputs, params, n_head, n_tokens_to_generate):
  from tqdm import tqdm
  for _ in tqdm(range(n_tokens_to_generate), "generating"):
    logits = gpt2(inputs, **params, n_head=n_head)
    next_id = np.argmax(logits[-1])
    inputs = np.append(inputs, [next_id])
  return list(inputs[len(inputs) - n_tokens_to_generate:])
```

```python
def mha(x, c_attn, c_proj, n_head):
  x = linear(x, **c_attn)
  qkv_heads = list(map(lambda x: np.split(x, n_head, axis=-1), np.split(x, 3, axis=-1)))
  causal_mask = (1 - np.tri(x.shape[0])) * -1e10
  out_heads = [attention(q, k, v, causal_mask) for q, k, v in zip(*qkv_heads)]
  x = linear(np.hstack(out_heads), **c_proj)
  return x

def main(prompt: str, n_tokens_to_generate: int = 40, model_size: str = "124M",
         models_dir: str = "models"):
  from utils import load_encoder_hparams_and_params
  encoder, hparams, params = load_encoder_hparams_and_params(model_size, models_dir)
  input_ids = encoder.encode(prompt)
  assert len(input_ids) + n_tokens_to_generate < hparams["n_ctx"]
  output_ids = generate(input_ids, params, hparams["n_head"], n_tokens_to_generate)
  output_text = encoder.decode(output_ids)
  return output_text
```
